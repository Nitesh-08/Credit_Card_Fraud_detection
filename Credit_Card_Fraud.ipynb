## Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.25 , random_state=42)
```

```python
print("X-train shape :", X_train.shape)
print("X-test shape :", X_test.shape)
print("y-train shape :", y_train.shape)
print("y-test shape :", X_test.shape)
```

## Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
#X_test = sc.transform(X_test)
```

```python
print("Training after standardised :\n" , X_train)
```

```python
import warnings
%matplotlib inline
pd.options.display.float_format = '{:.4f}'.format
np.set_printoptions(suppress=True)
warnings.filterwarnings('ignore')
pd.options.display.max_columns = None
```

```python
metric = pd.DataFrame(columns = ['Model','Accuracy','Precision','Recall','F1 Score'])
metric
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
print("Accuracy :",metrics.accuracy_score(y_test,y_pred))
print("Precision :",metrics.precision_score(y_test, y_pred))
print("Recall :",metrics.recall_score(y_test,y_pred))
print("F1-Score :",metrics.f1_score(y_test,y_pred))
metrics.plot_confusion_matrix(dt,X_test,y_test)
```

```python
metric.loc[len(metric)] = (['Decision Tree',metrics.accuracy_score(y_test,y_pred),metrics.precision_score(y_test,y_pred),metrics.recall_score(y_test,y_pred),metrics.f1_score(y_test,y_pred)])
metric
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print("Accuracy :", metrics.accuracy_score(y_test,y_pred))
print("Precision :",metrics.precision_score(y_test,y_pred))
print("Recall :",metrics.recall_score(y_test,y_pred))
print("F1-score :",metrics.f1_score(y_test,y_pred))
metrics.plot_confusion_matrix(knn,X_test,y_test)
```

```python
metric.loc[len(metric)] = (['KNN',metrics.accuracy_score(y_test,y_pred),metrics.precision_score(y_test,y_pred),metrics.recall_score(y_test,y_pred),metrics.f1_score(y_test,y_pred)])
metric
```

## HYPERPARAMETER TUNING

### Decision Tree

```python
from sklearn.model_selection import RandomizedSearchCV
grid_params = { 'criterion': ["gini", "entropy"],
                'max_depth': range(2,11),
                'min_samples_split': range(5,50,5),
                'min_samples_leaf': range(1,11),
                'ccp_alpha' : np.arange(0, 0.3, 0.03).tolist(),
               'random_state':[42]}

dt_gs = RandomizedSearchCV(DecisionTreeClassifier(), grid_params,verbose=1, n_jobs = -1,scoring='f1',cv=3)
dt_gs.fit(X_train, y_train)
dt_gs.best_params_
```

```python
dt_gs = DecisionTreeClassifier(criterion='entropy',max_depth=5,min_samples_split=5,min_samples_leaf=4,ccp_alpha=0.06,random_state=42)
dt_gs.fit(X_train,y_train)
y_pred = dt_gs.predict(X_test)
print("Accuracy :",metrics.accuracy_score(y_test,y_pred))
print("Precision :",metrics.precision_score(y_test,y_pred))
print("Recall :",metrics.recall_score(y_test,y_pred))
print("F1-score :",metrics.f1_score(y_test,y_pred))
metrics.plot_confusion_matrix(dt_gs,X_test,y_test)
```

```python
metric.loc[len(metric)] = (["DT Optimized",metrics.accuracy_score(y_test,y_pred),metrics.precision_score(y_test,y_pred),metrics.recall_score(y_test,y_pred),metrics.f1_score(y_test,y_pred)])
metric
```
